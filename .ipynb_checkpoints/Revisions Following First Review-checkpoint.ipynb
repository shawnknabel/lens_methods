# 11/30/19 The following code is to re-evaluate Spec's validity by differences between redshifts z1 and z2. z2 *should* be close to twice the value of z1 for the lensing effect to be prominent.
### This is true at low redshifts.

```python
spec_with_mass # pull up spec sample
```

### Plot z1 vs z2 for all Spec candidates

```python
fig, ax = plt.subplots(figsize=(8,8)) # square dimensions
X= spec_with_mass.z1
Y= spec_with_mass.z2
ax.plot(x, 2*x, 'k-') # line of slope 2
ax.plot(x, x, 'k--') # line of slope 1
#ax.plot(x, 2*x+0.15,'c:')
#ax.plot(x, 2*x-0.15,'c:')
plt.fill_between(x, 2*x+0.15, 2*x-0.15, # Fill region for "acceptable" (arbitrarily) error from 2:1 ratio
                 facecolor="cyan", # The fill color
                 color='cyan',       # The outline color
                 alpha=0.2)          # Transparency of the fill

plt.plot(X,Y, 'bo')
x = np.linspace(0, 1, 1000)
ax.set_xlim(0,0.8)
ax.set_ylim(0,0.8)
plt.xlabel("First Redshift Match - PG")
plt.ylabel("Second Redshift Match - ELG")
plt.title("GAMA Spectroscopy Redshift Matches")
plt.text(0.12,0.03, "This shaded region is +/- 0.15, chosen arbitrarily")
plt.text(0.4, 0.1, "For all candidates in this region: z1 > z2")
plt.text(0.55, 0.5,"This line is z1 = z2")
plt.text(0.35,0.675, "This line is z2 = 2 * z1")
plt.text(0.02, 0.65, "Here: z2 > 2*z1")
plt.savefig('/home/shawn/Desktop/gravitational_lensing_research/Lens_Project_Visuals/spec_z1_vs_z2.png')
plt.show()
```

### Cut spec sample to include only those candidates that appear in the shaded region above, +/- 0.15 from the 2:1 ratio.

```python
spec_zdiff_2pm015 = spec_with_mass[(spec_with_mass.z2 >= 2*spec_with_mass.z1-0.15) & (spec_with_mass.z2 <= 2*spec_with_mass.z1+0.15)]
```

```python
#spec_zdiff_2pm015 = spec_zdiff_2pm015.reset_index()
#spec_zdiff_2pm015 = spec_zdiff_2pm015.drop(columns = ["index","Unnamed: 0","Unnamed: 0.1"])
```

### Calculations for maximum masses within the 2 arcsec aperture given the lens and source redshifts

```python
z1 = spec_zdiff_2pm015.z1
z2 = spec_zdiff_2pm015.z2

max_masses = z1*z2*10**18/(z1+z2)
```

```python
max_masses
```

### Insert additional column of "max_mass" using values calculated above.

```python
spec_zdiff_2pm015.insert(8, 'max_mass', max_masses, allow_duplicates = False)
spec_zdiff_2pm015
```
