```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from astropy import constants as const
from astropy.io import fits
c = const.c.to('km/s').value #c in km/s
h0 = 73.8 #h0 in km/(Mpc*s) Riess2011
from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(h0,Om0=0.262)
```

```python
spec = pd.read_csv('/home/shawn/Desktop/gravitational_lensing_research/Lens_Project_Files/spec_latest.csv')
mac = pd.read_csv('/home/shawn/Desktop/gravitational_lensing_research/Lens_Project_Files/mac_latest.csv')
zoo = pd.read_csv('/home/shawn/Desktop/gravitational_lensing_research/Lens_Project_Files/zoo_latest.csv')
```

```python
# import lambdarstellarmasses
# Opening GAMA MagPhys data from G09, G12, and G15 (DR3)
hdul = fits.open('/home/shawn/Desktop/gravitational_lensing_research/FITS_Files/StellarMassesLambdar.fits')  # open a FITS file
hdul.verify('fix')
masses = hdul[1].data  # assume the first extension is a table
print(masses.columns)
```

```python
# create dataframe of masses objects to compare to candidates
GAMA_ID = masses.CATAID.byteswap().newbyteorder()
logmstar = masses.logmstar.byteswap().newbyteorder()
logmintsfh = masses.logmintsfh.byteswap().newbyteorder()
logmremnants = masses.logmremnants.byteswap().newbyteorder()
fluxscale = masses.fluxscale.byteswap().newbyteorder()

#masses_list = list(zip(GAMA_ID, logmstar, logmintsfh, logmremnants, fluxscale))

#masses_list
lambdar_masses = pd.DataFrame(
    {'GAMA_ID' : GAMA_ID,
     'logmstar' : logmstar, 
     'logmintsfh' : logmintsfh, 
     'logmremnants' : logmremnants, 
     'fluxscale' : fluxscale})
lambdar_masses.GAMA_ID = lambdar_masses.GAMA_ID.astype(int)
lambdar_masses
```

```python
spec_lambdar = pd.merge(
    spec, 
    lambdar_masses, 
    how = 'left', 
    on = 'GAMA_ID')

#print(lambdar_masses.fluxscale)
print(np.mean((10**(spec_lambdar.logmstar))/spec_lambdar.STELLAR_MASS))
print(np.mean(10**(spec_lambdar.logmintsfh)/spec_lambdar.STELLAR_MASS))

plt.hist(np.log10(spec_lambdar.STELLAR_MASS), 
         histtype='step',
        label='magphys')
plt.hist(spec_lambdar.logmstar,
         histtype='step',
        label='lambdar aperture mass')
plt.hist(spec_lambdar.logmintsfh,
        histtype='step',
        label='int sfh')
plt.legend()
```

# MagPhys masses (calcualted by sfh) for Spec candidates are smaller than Lambdar sfh by an average of ~ 47%, and I haven't even corrected for the aperture. I'm not sure if Lambdar sfh needs to be corrected. I don't think so.

```python
# correct logmstar
spec_lambdar['logmstar_total'] = 0.0
for i in range(len(spec_lambdar.GAMA_ID)):
    if spec_lambdar.fluxscale[i] >= 1.0:
        spec_lambdar['logmstar_total'][i] = spec_lambdar.logmstar[i] + np.log10(spec_lambdar.fluxscale[i]) - 2 * np.log10(h0/70)
    elif (spec_lambdar.fluxscale[i] >= 0.0) & (spec_lambdar.fluxscale[i] < 1.0):
        spec_lambdar['logmstar_total'][i] = spec_lambdar.logmstar[i]            
    else: spec_lambdar['logmstar_total'][i] = 0.0
spec_lambdar.logmstar_total
```

```python
# check to see how this compares to the other masses
print(np.mean(10**(spec_lambdar.logmstar_total)/spec_lambdar.STELLAR_MASS))

plt.hist(np.log10(spec_lambdar.STELLAR_MASS), 
         histtype='step',
        label='magphys sfh')
#plt.hist(spec_lambdar.logmstar,
#         histtype='step',
#        label='lambdar')
plt.hist(spec_lambdar.logmstar_total[spec_lambdar.logmstar_total != 0],
        histtype='step',
        label='lambdar total mstar')
plt.hist(spec_lambdar.logmintsfh,
        histtype='step',
        label='lambdar int sfh')
plt.xlabel('$log(M_*/M_{\odot})$')
plt.ylabel('count')
plt.title('GAMA spectroscopy candidates stellar mass measurements')
plt.legend()
```
