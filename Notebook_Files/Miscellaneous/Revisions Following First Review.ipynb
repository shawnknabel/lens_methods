## I want to see what the estimated einstein rings look like for the ratios of z2 to z1
Estimating dark matter fraction of 0.31 - from Barnabe et al. 2011 (SLACS, assuming Salpeter IMF)

```python
spec_with_mass.dtypes
```

```python
# assuming point mass for first approximation
M = spec_with_mass.STELLAR_MASS/0.69 #total mass, baryonic and dark matter, to be taken to 1/2 power
z1 = spec_with_mass.z1
z2 = spec_with_mass.z2

A = (M/(10**11.09))**(1/2) # mass term
B = ((c/(h0*1000))*(z1*z2/(z2-z1)))**(-1/2) # distances term, h0*1000 gives km/Gpc*s, removes all candidate where z2<z1

theta_e = A*B #gives arcsecs
#theta_e_deg = theta_e/3600 # gives theta_e in degrees

theta_e
```
